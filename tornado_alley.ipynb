```python
import pandas as pd
import datetime as dt
from pathlib import Path
```

```python
# Loading data
# I have the renamed csv in a seperate 'Resources' folder so you may need to change your path
file_path = Path("../Resources/tornado_data.csv")
tornado_df = pd.read_csv(file_path)
tornado_df.head()
```

```python
# Find Null Columns
tornado_df.isnull().sum()
```

```python
# Rename the Existing DataFrame Columns
tornado_df.rename(columns={'yr': 'Year', 'mo': 'Month', 'dy': 'Day', 'date': 'Date', 'st': 'State', 'mag': 'Magnitude', 'inj': 'Injuries', 'fat': 'Fatalities', 'slat': 'Start Lat', 'slon': 'Start Long', 'elat': 'Ending Lat', 'elon': 'Ending Long', 'len': 'Length in Miles', 'wid': 'Width in Yards'}, inplace=True)
tornado_df.head()
```

```python
# Export renamed csv
# May need to change your path
tornado_df.to_csv("../Resources/renamed_tornado_data.csv", index=False)
```
